```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
#首先定义函数
def f(x, y):
    return 2.0*pow(x,2)+5.0*pow(y,2)
#定义x，y的取值范围
x = np.linspace(-12,12,300)
y = np.linspace(-6,6,300)
#组成取值范围网格
theta1,theta2= np.meshgrid(x, y)
jtheta= f(theta1,theta2)
```

```python
ax = plt.axes(projection='3d')
ax.contour3D(theta1,theta2,jtheta,200)
ax.set_xlabel('theta1')
ax.set_ylabel('theta2')
ax.set_zlabel('jtheta')
#调整观察角度和方位角。这里将俯仰角设为60度，把方位角调整为35度
ax.view_init(10, 30)
```

```python
#计算梯度下降
def gradient(a,theta1,theta2):
    thetas=np.array([theta1,theta2]).reshape(2,1)
    #函数变量的系数
    params=np.array([4,10]).reshape(1,2)
    theta1s=[]
    theta2s=[]
    jthetas=[]
    #梯度
    gradient=np.dot(params,thetas)
    print(gradient)
    while not np.all(np.absolute(gradient)<=0.00001):
        theta1s.append(thetas[0,0])
        theta2s.append(thetas[1,0])
        jthetas.append(2*pow(thetas[0,0],2)+5*pow(thetas[1,0],2))
        thetas=thetas-a*np.dot(params,thetas)
        gradient=np.dot(params,thetas)
        print(gradient)
    return np.array(theta1s),np.array(theta2s),np.array(jthetas)
```

```python
theta1s,theta2s,jthetas=gradient(0.03,12,6)
```

```python
#梯度下降可视化
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(theta1,theta2,jtheta,20)
ax.set_xlabel('theta1')
ax.set_ylabel('theta2')
ax.set_zlabel('jtheta')
ax.view_init(0,75)
ax.plot3D(theta1s,theta2s,jthetas,'rx-')
```
