```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,1000)
print(type(x))
```

```python
y=2*(x)*x-x+3
```

```python
print(x.shape,y.shape)
```

```python
plt.plot(x,y)
```

```python
#梯度下降最简易版
def getPoints(alpha,theta):
    thetas=[]
    Js=[]
    while abs(4*theta-1)>=0.01:
        #计算函数值
        js=2*pow(theta,2)-theta+3
        thetas.append(theta)
        Js.append(js)
        #更新theta的操作
        theta=theta-alpha*(4*theta-1)
    return thetas,Js
thetas1,Js1=getPoints(0.3,-5)
thetas2,Js2=getPoints(0.1,-5)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.plot(x,y)
plt.plot(thetas1,Js1,'rx-')
plt.legend(labels=['Js','alpha=0.2'])
plt.subplot(122)
plt.plot(x,y)
plt.plot(thetas2,Js2,'rx-')
plt.legend(labels=['Js','alpha=0.1'])
```
