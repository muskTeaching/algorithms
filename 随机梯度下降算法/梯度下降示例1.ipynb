```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,1000)
print(type(x))
```

```python
y=2*x.T*x-x+3
```

```python
print(x.shape,y.shape)
```

```python
plt.plot(x,y)
```

```python
def getPoints(a,s):
    ss=[]
    jss=[]
    while abs(4*s-1)>0.01:
        js=2*pow(s,2)-s+3
        ss.append(s)
        jss.append(js)
        s=s-a*(4*s-1)
    return ss,jss
ss1,jss1=getPoints(0.3,-5)
ss2,jss2=getPoints(0.1,-5)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.plot(x,y)
plt.plot(ss1,jss1,'rx-')
plt.legend(labels=['origin','a=0.2'])
plt.subplot(122)
plt.plot(x,y)
plt.plot(ss2,jss2,'rx-')
plt.legend(labels=['origin','a=0.1'])
```
